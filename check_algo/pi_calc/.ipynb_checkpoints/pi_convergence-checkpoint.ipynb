```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df0 = pd.read_csv("pi_converg.csv", index_col = 0)
fig, ax = plt.subplots(nrows = 1, ncols = 1)
df0.astype(float).plot(ax=ax, ylim=[0.75,0.85])
plt.hlines(0.785, xmin=0, xmax=10000)
plt.grid()
```

```python
dfk = pd.read_csv("pi_sqrtK.csv")
K = dfk.astype(float).iloc[:,0] 
pi= dfk.astype(float).iloc[:,1] 
plt.plot(K, pi)
plt.xlim(0.9,0.08)
plt.hlines(0.785, xmin=0, xmax=10000)

plt.grid()
```
