```python
import pandas as pd
import matplotlib.pyplot as plt
df0 = pd.read_csv("pi_converg0.dat", index_col = 0)
df1 = pd.read_csv("pi_converg1.dat", index_col = 0)
df2 = pd.read_csv("pi_converg2.dat", index_col = 0)
fig, ax = plt.subplots(nrows = 1, ncols = 1)
df0.astype(float).plot(ax=ax, ylim=[0.75,0.85])
df1.astype(float).plot(ax=ax, ylim=[0.75,0.85])
df2.astype(float).plot(ax=ax, ylim=[0.75,0.85])
plt.grid()
```
